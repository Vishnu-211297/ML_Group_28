**PART 2 - DATA PRE-PROCESSING**

```python
from imblearn.over_sampling import ADASYN
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Data/IBM HR Analytics.csv')
```

```python
df.head()
```

```python
print('Shape: ',df.shape)
```

DROP IRRELEVANT COLUMNS

```python
df = df.drop(columns = ['EmployeeCount', 'EmployeeNumber', 'Over18', 'StandardHours'])
```

```python
X = df.drop('Attrition', axis=1)
y = df['Attrition']
```

DUMMYING THE CATEGORICAL FEATURES

```python
df = pd.get_dummies(X, drop_first=False)
```

```python
# Separate the dummy variables from the numerical variables
dummy_cols = df.select_dtypes(include='uint8').columns
num_cols = list(set(df.select_dtypes(include='int64').columns) - set(dummy_cols))

# Standardize the numerical variables
scaler = StandardScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])

# Combine the standardized numerical variables with the dummy variables
df_std = pd.concat([df[num_cols], df[dummy_cols]], axis=1)
```

ADDRESSING THE CLASS IMBALANCE ISSUE

```python
#Use ADASYN to resolve class imbalance
ada = ADASYN(random_state=42)
X_resampled, y_resampled = ada.fit_resample(df_std, y)
```

**PART 3 - ANALYSIS**

```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Data/Data_Scaled.csv")
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
from sklearn.ensemble import VotingClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop("Attrition", axis=1), df["Attrition"], test_size=0.2, random_state=42)
```

**K-NEAREST NEIGHBOR**

```python
param_grid = {'n_neighbors': [3, 5, 7], 'weights': ['uniform', 'distance']}
```

```python
knn = KNeighborsClassifier(metric='manhattan')
```

```python
knn_gs = GridSearchCV(knn, param_grid, cv=5)
```

```python
knn_gs.fit(X_train, y_train)
```

```python
knn_pred = knn_gs.predict(X_test)
```

**XGBOOST**

```python
gbm = GradientBoostingClassifier()
```

```python
gbm_cv = cross_val_score(gbm, X_train, y_train, cv=5)
```

```python
gbm.fit(X_train, y_train)
```

```python
gbm_pred = gbm.predict(X_test)
```

**SUPPORT VECTOR MACHINE**

```python
svm = SVC(probability=True)
```

```python
svm_param_grid = {'kernel': ['linear'],
                   'C': [0.01, 0.1],
                   'gamma': [0.01, 0.1]}
```

```python
svm_grid = GridSearchCV(svm, svm_param_grid, cv=5)
```

```python
svm_grid.fit(X_train, y_train)
```

```python
svm_pred = svm_grid.predict(X_test)
```

**RANDOM FOREST**

```python
rf = RandomForestClassifier(n_estimators=100)
```

```python
rf_param_grid = {'n_estimators': [50, 100, 200],
                 'max_depth': [5, 10, None]}
```

```python
rf_grid = GridSearchCV(rf, rf_param_grid, cv=5)
```

```python
rf.fit(X_train, y_train)
```

```python
rf_pred = rf.predict(X_test)
```

**MODEL EVALUATION**

```python
knn_acc = accuracy_score(y_test, knn_pred)
svm_acc = accuracy_score(y_test, svm_pred)
rf_acc = accuracy_score(y_test, rf_pred)
gbm_acc = accuracy_score(y_test, gbm_pred)
print("KNN Accuracy:", knn_acc)
print("SVM Accuracy:", svm_acc)
print("Random Forest Accuracy:", rf_acc)
print("Gradient Boosting Accuracy:", gbm_acc)
```

```python
knn_precision = precision_score(y_test, knn_pred)
svm_precision = precision_score(y_test, svm_pred)
rf_precision = precision_score(y_test, rf_pred)
gbm_precision = precision_score(y_test, gbm_pred)
print("KNN Precision:", knn_precision)
print("SVM Precision:", svm_precision)
print("Random Forest Precision:", rf_precision)
print("Gradient Boosting Precision:", gbm_precision)
```

```python
knn_acc = recall_score(y_test, knn_pred)
svm_acc = recall_score(y_test, svm_pred)
rf_acc = recall_score(y_test, rf_pred)
gbm_acc = recall_score(y_test, gbm_pred)
print("KNN Recall:", knn_acc)
print("SVM Recall:", svm_acc)
print("Random Recall:", rf_acc)
print("Gradient Recall:", gbm_acc)
```

```python
knn_acc = f1_score(y_test, knn_pred)
svm_acc = f1_score(y_test, svm_pred)
rf_acc = f1_score(y_test, rf_pred)
gbm_acc = f1_score(y_test, gbm_pred)
print("KNN f1_score:", knn_acc)
print("SVM f1_score:", svm_acc)
print("Random f1_score:", rf_acc)
print("Gradient f1_score:", gbm_acc)
```

**PLOTTING THE CONFUSION MATRIX**

```python
confusion_knn = confusion_matrix(y_test, knn_pred)
confusion_svm = confusion_matrix(y_test, svm_pred)
confusion_rf = confusion_matrix(y_test, rf_pred)
confusion_gbm = confusion_matrix(y_test, gbm_pred)
```

```python
# Plot the confusion matrix
sns.heatmap(confusion_rf, annot=True, cmap='Blues', fmt='g')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix Random Forest')
plt.show()
```

```python
# Plot the confusion matrix
sns.heatmap(confusion_gbm, annot=True, cmap='Blues', fmt='g')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix XGBoost')
plt.show()
```
